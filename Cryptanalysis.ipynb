```python
text = input("Enter cipher text: ").upper()
tc = len(text)
print("Total characters:", tc)

ec = {}
for char in text:
    if char in ec:
        ec[char] += 1
    else:
        ec[char] = 1

char_array = list(ec.items())

def get_count(item):
    return item[1]

char_array.sort(key=get_count, reverse=True)
top_three = char_array[:3]
print(top_three)

percentage_array = []
for char, count in top_three:
    percentage = (count / tc) * 100
    percentage_array.append((char, round(percentage, 2)))
print(percentage_array)


shift = int(input("Enter shift value for decryption: "))

def decrypt_caesar(text, shift):
    result = ''
    for char in text:
        if char.isalpha():
            decrypted_char = chr((ord(char) - ord('A') - shift) % 26 + ord('A'))

            result += decrypted_char
        else:
            result += char
    return result

decrypted_text = decrypt_caesar(text, shift)
print("Decrypted text:", decrypted_text)

```
